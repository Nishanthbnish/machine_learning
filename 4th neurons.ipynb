```python
import numpy as np
x=np.array(([2,9],[1,5],[3,6]),dtype=float)
y=np.array(([89],[92],[86]),dtype=float)
x=x/np.amax(x,axis=0)
y=y/100

def sigmoid(x):
    return 1/(1+ np.exp(-x))
def sigmoid_grad(x):
    return x*(1-x)

epoch=1000
eta=0.2
input_neuron=2
hidden_neuron=3
output_neuron=1

wh=np.random.uniform(size=(input_neuron,hidden_neuron))
bh=np.random.uniform(size=(1,hidden_neuron))
wout=np.random.uniform(size=(hidden_neuron,output_neuron))
bout=np.random.uniform(size=(1,output_neuron))

for i in range(epoch):
     #forward formation
    h_ip=np.dot(x,wh)+bh
    h_act = sigmoid(h_ip)
    o_ip=np.dot(h_act,wout)
    output = sigmoid(o_ip)
    
    #error in output
    eo=y-output
    outgrad=sigmoid_grad(output)
    d_output=eo*outgrad
    
    #error in hidden 
    eh=d_output.dot(wout.T)
    hiddengrad=sigmoid_grad(h_act)
    d_hidden= eh*hiddengrad
    wout += h_act.T.dot(d_output)*eta   
    wh += x.T.dot(d_hidden)*eta
    
    print("Normal Input:\n", str(x))
    print("Actual ouput:\n", str(y))
    print("predicted output:\n",output)
   
    
```
